```python
from pathlib import Path

try:
    ROOT = Path(__file__).parent.parent
except NameError:
    ROOT = Path.cwd().parent

DATA_DIR = ROOT / "Data"
SRC_DIR  = DATA_DIR / "BTC_data"
OUT_FILE = DATA_DIR / "BTCUSDT_1m.parquet"
```

```python
import pandas as pd

import matplotlib.pyplot as plt

df = pd.read_parquet("BTCUSDT_1m.parquet")

# convert time column if present
time_col = next((c for c in df.columns if "time" in c.lower() or "date" in c.lower()), None)
if time_col is not None:
    df[time_col] = pd.to_datetime(df[time_col], unit="ms", errors="ignore")
    df.set_index(time_col, inplace=True)

# plot close price
df["close"].plot(title="BTCUSDT 1m Close Price", figsize=(12, 6))
plt.xlabel("Time")
plt.ylabel("Close Price")
plt.grid(True)
plt.show()
```
